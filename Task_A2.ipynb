```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch
```

## 1. Load `gpt2` using HuggingFace Transformers.

```python
model_name = "gpt2"

tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

model.eval()  # set to evaluation mode
```

```python
def generate_text(prompt, max_length=50):
    inputs = tokenizer.encode(prompt, return_tensors="pt")

    outputs = model.generate(
        inputs,
        max_length=max_length,
        do_sample=True,
        top_k=50,
        temperature=0.2,
        num_return_sequences=1,
        attention_mask=torch.ones_like(inputs),
        pad_token_id=tokenizer.eos_token_id
    )

    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

## 2. Generate text for each of the following 3 prompts using `top_k=50` and `temperature=0.8`:   
- Prompt 1: "The future of artificial intelligence is"
 - Prompt 2: "In a transformer model, the attention mechanism"
- Prompt 3: "The most important skill for a data engineer is"

```python
prompts = [
    "The future of artificial intelligence is",
    "In a transformer model, the attention mechanism",
    "The most important skill for a data engineer is"
]

for i, prompt in enumerate(prompts, 1):
    print(f"\n--- Output {i} ---")
    print(generate_text(prompt))
```

### Effect of Changing Temperature from 0.8 to 0.2

Reducing the temperature from 0.8 to 0.2 makes the output more deterministic and less random. At lower temperature, the model strongly prefers high-probability tokens, resulting in safer and more repetitive text. Creativity and diversity decrease, and the responses become more predictable. In contrast, a temperature of 0.8 allows more variation and exploration in generated text.
